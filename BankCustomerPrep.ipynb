# Data Import & QA 

### Merged customer information with account details
### Removed duplicate entries while preserving data integrity
### Standardized geographical categories

```python
import pandas as pd
churn_cust_info = pd.read_excel("Bank_Churn_Messy.xlsx")
churn_cust_info.head()
```

```python
churn_acct_info = pd.read_excel("Bank_Churn_Messy.xlsx", sheet_name = 1)
churn_acct_info.head()
```

```python
churn_df = churn_cust_info.merge(churn_acct_info, how = 'left', on = 'CustomerId')
churn_df.head()
```

```python
churn_df.info()
```

```python
churn_df = churn_df.drop('Tenure_y', axis=1)

churn_df = churn_df.rename({'Tenure_x': 'Tenure'}, axis=1)
```

```python
churn_df.head()
```

# Data Cleaning

### Handled missing values using appropriate statistical methods
### Standardized currency formats across financial fields
### Created consistent categorical variables

```python
churn_df.duplicated(keep=False)
```

```python
churn_df.tail()
```

```python
churn_df = churn_df.drop_duplicates()
```

```python
churn_df["Balance"] = churn_df["Balance"].str.replace("€","").astype("float")
churn_df.info()
```

```python
churn_df["Balance"] = churn_df["Balance"].astype(str).replace("€","").astype(float)

churn_df.info()
```

```python
churn_df[churn_df.isna().any(axis=1) == True]
```

```python
churn_df = churn_df.fillna(value={'Surname': 'MISSING','Age': churn_df['Age'].median()})
```

```python
churn_df.info()
```

```python
churn_df.iloc[[28,1221,9389]]
```

```python
churn_df.describe()
```

```python
churn_df['EstimatedSalary'] = churn_df['EstimatedSalary'].str.replace('€', '').astype(float)

churn_df['EstimatedSalary'] = churn_df['EstimatedSalary'].replace(-999999, churn_df['EstimatedSalary'].median())

print(churn_df.describe())
```

```python
churn_df['Geography'].value_counts()
```

```python
import numpy as np

churn_df['Geography'] = np.where(churn_df['Geography'].isin(['FRA','France','French']), 'France', churn_df['Geography'])

churn_df['Geography'].value_counts()
```

# Data Prep & Feature Engineering

```python
modelling_df = churn_df.drop(['CustomerId','Surname'], axis=1)

modelling_df.head()
```

```python
modelling_df= pd.get_dummies(modelling_df, drop_first=True)
```

```python
modelling_df['Balance_vs_Sal'] = modelling_df['Balance']/modelling_df['EstimatedSalary']

modelling_df.head()
```
